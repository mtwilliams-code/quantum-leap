```python
from IPython.display import display
import pdfplumber
import re

# ---- Config ----
pdf_path = "air_force_budget.pdf"
TOP_N = 10                 # how many biggest numbers to show
START_PAGE = 1             # 1-based inclusive
END_PAGE = None            # 1-based inclusive; None = end of document
X_TOL = 1.0                # pdfplumber word extraction x_tolerance
Y_TOL = 1.0                # pdfplumber word extraction y_tolerance
DRAW_ALL_WORDS = True      # also draw all words in light gray for context
RESOLUTION = 150           # page render dpi

# ---- Number parsing (naive; no scaling) ----
_GROUP_SEP = ",\u00A0\u2009"  # comma, NBSP, thin space
NUM_RE = re.compile(
    r"^\s*"
    r"(?P<prefix>[\$\(]?)\s*"
    r"(?P<num>(?:\d{1,3}(?:[" + _GROUP_SEP + r"]\d{3})+|\d+)(?:\.\d+)?|(?:\.\d+))"
    r"\s*(?P<suffix>[%\)]?)"
    r"(?P<foot>[\*\u2020\u2021\u00B9\u00B2\u00B3\u2070-\u2079]*)\s*$"
)

def parse_number(tok: str):
    tok = (tok or "").strip()
    m = NUM_RE.match(tok)
    if not m:
        return None
    num = m.group("num").replace(",", "").replace("\u00A0", "").replace("\u2009", "")
    try:
        val = float(num)
    except ValueError:
        return None
    if m.group("prefix") == "(" and m.group("suffix") == ")":
        val = -val
    return val

# ---- Collect candidates across pages ----
hits = []
with pdfplumber.open(pdf_path) as pdf:
    total_pages = len(pdf.pages)
    start_idx = max(1, START_PAGE)
    end_idx = total_pages if END_PAGE is None else min(END_PAGE, total_pages)

    for page_num in range(start_idx, end_idx + 1):
        page = pdf.pages[page_num - 1]
        words = page.extract_words(x_tolerance=X_TOL, y_tolerance=Y_TOL) or []
        page_text = page.extract_text() or ""
        try:
            tables = page.find_tables() or []
        except Exception:
            tables = []

        for w in words:
            tok = str(w.get("text", ""))
            val = parse_number(tok)
            if val is None:
                continue
            bbox = (float(w.get("x0", 0)), float(w.get("top", 0)),
                    float(w.get("x1", 0)), float(w.get("bottom", 0)))
            hits.append({
                "value": val,
                "raw": tok,
                "page_num": page_num,
                "bbox": bbox,
                "word": w,
                "words_count": len(words),
                "tables_count": len(tables),
                "page_bbox": page.bbox,
            })

# ---- Sort and show top N ----
hits.sort(key=lambda h: h["value"], reverse=True)
top_hits = hits[:TOP_N]

print(f"Scanned pages {START_PAGE}..{END_PAGE or total_pages}. Found {len(hits)} numeric tokens.")
print(f"Showing top {len(top_hits)} hits (no scaling applied).")

with pdfplumber.open(pdf_path) as pdf:
    for rank, h in enumerate(top_hits, 1):
        page = pdf.pages[h["page_num"] - 1]
        words = page.extract_words(x_tolerance=X_TOL, y_tolerance=Y_TOL) or []

        print(
            f"#{rank} value={h['value']} raw='{h['raw']}'  "
            f"page={h['page_num']}  bbox={tuple(round(x,1) for x in h['bbox'])}  "
            f"words={h['words_count']} tables={h['tables_count']} page_bbox={page.bbox}"
        )

        im = page.to_image(resolution=RESOLUTION)
        if DRAW_ALL_WORDS and words:
            im.draw_rects(words, stroke="lightgray", stroke_width=1)

        x0, top, x1, bottom = h["bbox"]
        im.draw_rects([{"x0": x0, "top": top, "x1": x1, "bottom": bottom}],
                      stroke="red", stroke_width=2)

        display(im)
```

```python
from IPython.display import display
import pdfplumber
import re

from conductor_takehome.scale import detect_scale_phrase, scale_factor

# ---- Config ----
pdf_path = "air_force_budget.pdf"
START_PAGE = 1             # 1-based inclusive
END_PAGE = None            # 1-based inclusive; None = end of doc
TOP_N = 10                 # show top N by scaled value
X_TOL = 1.0                # pdfplumber word extraction x_tolerance
Y_TOL = 1.0                # pdfplumber word extraction y_tolerance
DRAW_ALL_WORDS = True      # draw all words for context
RESOLUTION = 150           # page render dpi

# ---- Number parsing (handles $, commas, decimals, (.negative), %, footnotes) ----
_GROUP_SEP = ",\u00A0\u2009"  # comma, NBSP, thin space
NUM_RE = re.compile(
    r"^\s*"
    r"(?P<prefix>[\$\(]?)\s*"
    r"(?P<num>(?:\d{1,3}(?:[" + _GROUP_SEP + r"]\d{3})+|\d+)(?:\.\d+)?|(?:\.\d+))"
    r"\s*(?P<suffix>[%\)]?)"
    r"(?P<foot>[\*\u2020\u2021\u00B9\u00B2\u00B3\u2070-\u2079]*)\s*$"
)

def parse_number(tok: str):
    tok = (tok or "").strip()
    m = NUM_RE.match(tok)
    if not m:
        return None
    num = (
        m.group("num")
        .replace(",", "")
        .replace("\u00A0", "")
        .replace("\u2009", "")
    )
    try:
        val = float(num)
    except ValueError:
        return None
    if m.group("prefix") == "(" and m.group("suffix") == ")":
        val = -val
    return val

# ---- Collect hits across pages, applying page-level scale ----
hits = []
with pdfplumber.open(pdf_path) as pdf:
    total_pages = len(pdf.pages)
    start_idx = max(1, START_PAGE)
    end_idx = total_pages if END_PAGE is None else min(END_PAGE, total_pages)

    for page_num in range(start_idx, end_idx + 1):
        page = pdf.pages[page_num - 1]
        words = page.extract_words(x_tolerance=X_TOL, y_tolerance=Y_TOL) or []
        try:
            tables = page.find_tables() or []
        except Exception:
            tables = []

        # Detect page-level scale phrase once per page
        scl_name, scl_phrase, scl_bbox = detect_scale_phrase(words)
        factor = scale_factor(scl_name)

        for w in words:
            tok = str(w.get("text", ""))
            raw_val = parse_number(tok)
            if raw_val is None:
                continue
            bbox = (
                float(w.get("x0", 0)), float(w.get("top", 0)),
                float(w.get("x1", 0)), float(w.get("bottom", 0))
            )
            hits.append({
                "page_num": page_num,
                "raw": tok,
                "raw_value": raw_val,
                "scaled_value": raw_val * factor,
                "bbox": bbox,
                "scale_name": scl_name,
                "scale_phrase": scl_phrase,
                "scale_bbox": scl_bbox,
                "words_count": len(words),
                "tables_count": len(tables),
                "page_bbox": page.bbox,
            })

# ---- Sort and render top N by scaled value ----
hits.sort(key=lambda h: h["scaled_value"], reverse=True)
top_hits = hits[:TOP_N]

print(f"Scanned pages {START_PAGE}..{END_PAGE or total_pages}. "
      f"Found {len(hits)} numeric tokens. Showing top {len(top_hits)} by scaled value.")

with pdfplumber.open(pdf_path) as pdf:
    for rank, h in enumerate(top_hits, 1):
        page = pdf.pages[h["page_num"] - 1]
        words = page.extract_words(x_tolerance=X_TOL, y_tolerance=Y_TOL) or []

        print(
            f"#{rank} scaled={h['scaled_value']:,} raw={h['raw_value']} "
            f"('{h['raw']}') page={h['page_num']} "
            f"bbox={tuple(round(x,1) for x in h['bbox'])} "
            f"scale={h['scale_name'] or 'none'} phrase={h['scale_phrase']!r} "
            f"words={h['words_count']} tables={h['tables_count']}"
        )

        im = page.to_image(resolution=RESOLUTION)

        # Draw all words light gray (context)
        if DRAW_ALL_WORDS and words:
            im.draw_rects(words, stroke="lightgray", stroke_width=1)

        # Draw scale phrase bbox in blue (if found)
        if h["scale_bbox"]:
            x0, top, x1, bottom = h["scale_bbox"]
            im.draw_rects(
                [{"x0": x0, "top": top, "x1": x1, "bottom": bottom}],
                stroke="blue", stroke_width=2
            )

        # Draw the number bbox in red
        x0, top, x1, bottom = h["bbox"]
        im.draw_rects(
            [{"x0": x0, "top": top, "x1": x1, "bottom": bottom}],
            stroke="red", stroke_width=2
        )

        display(im)
```

```python
from IPython.display import display
import pdfplumber
import re

from conductor_takehome.scale import (
    detect_scale_phrase,
    detect_scale_phrase_in_region,
    scale_factor,
)

# -------- Config --------
pdf_path = "air_force_budget.pdf"
START_PAGE = 1
END_PAGE = None
TOP_N = 12
X_TOL = 1.0
Y_TOL = 1.0
RESOLUTION = 150
DRAW_ALL_WORDS = False
# Thresholds (set to a number or leave as None)
MAX_SCALED = None      # e.g., 10_000_000
MIN_SCALED = None      # e.g., 1_000_000
MAX_RAW = None         # e.g., 100_000
MIN_RAW = None         # e.g., 1_000


# -------- Number parsing (no scaling) --------
_GROUP_SEP = ",\u00A0\u2009"  # comma, NBSP, thin space
NUM_RE = re.compile(
    r"^\s*(?P<prefix>[\$\(]?)\s*"
    r"(?P<num>(?:\d{1,3}(?:[" + _GROUP_SEP + r"]\d{3})+|\d+)(?:\.\d+)?|(?:\.\d+))"
    r"\s*(?P<suffix>[%\)]?)"
    r"(?P<foot>[\*\u2020\u2021\u00B9\u00B2\u00B3\u2070-\u2079]*)\s*$"
)
def parse_number(tok: str):
    tok = (tok or "").strip()
    m = NUM_RE.match(tok)
    if not m:
        return None
    num = (
        m.group("num")
        .replace(",", "")
        .replace("\u00A0", "")
        .replace("\u2009", "")
    )
    try:
        val = float(num)
    except ValueError:
        return None
    if m.group("prefix") == "(" and m.group("suffix") == ")":
        val = -val
    if m.group("suffix") == "%":
        return None  # drop percents from ranking
    return val

# -------- Row/unit classification --------
HEADCOUNT_PAT = re.compile(
    r"\b(end\s*strength|work[-\s]*years?|workyears?|fte|headcount|person(n)?el|items\s*managed|quantity|issues|receipts|requisitions|contracts|units?)\b",
    re.I,
)
MONEY_HINT_PAT = re.compile(
    r"\$|\b(?:usd|dollars?)\b|\(\s*\$?\s*(?:in\s+)?(?:K|M|B)\s*\)",
    re.I,
)


def classify_units_for_word(word: dict, line_words: list[dict]) -> str:
    """
    Return 'people', 'money', or 'unknown' based on the left-side label/context.
    Heuristic: look at words left of the number on the same visual line.
    """
    x_left = float(word.get("x0", 0.0))
    # words roughly on the same line (same y bucket) provided by caller
    left_text = " ".join(
        str(w.get("text", "")) for w in line_words if float(w.get("x1", 0.0)) <= x_left - 2.0
    )
    if not left_text:
        return "unknown"
    if HEADCOUNT_PAT.search(left_text):
        return "people"
    if MONEY_HINT_PAT.search(left_text):
        return "money"
    # Weak heuristic: rows with decimals are more often money when table states ($M)
    # but don't force; leave as unknown to use table/page scale
    return "unknown"

def inside(box, region, tol=0.5):
    x0,y0,x1,y1 = box
    rx0,ry0,rx1,ry1 = region
    return (x0 >= rx0 - tol) and (x1 <= rx1 + tol) and (y0 >= ry0 - tol) and (y1 <= ry1 + tol)

def lines_by_y(words, bucket=10.0):
    lines = {}
    for w in words:
        y = round(float(w.get("top", 0.0)) / bucket) * bucket
        lines.setdefault(y, []).append(w)
    return lines

# -------- Collect hits with table-scoped scale + row unit classification --------
hits = []
with pdfplumber.open(pdf_path) as pdf:
    total_pages = len(pdf.pages)
    start_idx = max(1, START_PAGE)
    end_idx = total_pages if END_PAGE is None else min(END_PAGE, total_pages)

    for page_num in range(start_idx, end_idx + 1):
        page = pdf.pages[page_num - 1]
        words = page.extract_words(x_tolerance=X_TOL, y_tolerance=Y_TOL) or []
        line_map = lines_by_y(words)

        # Tables and table-specific scale
        try:
            tables = page.find_tables() or []
        except Exception:
            tables = []
        table_regions = [t.bbox for t in tables if getattr(t, "bbox", None)]
        table_scales = []
        for tb in table_regions:
            tn, tp, tbbox = detect_scale_phrase_in_region(words, tb)
            table_scales.append((tb, tn, tp, tbbox))

        # Page fallback scale
        p_name, p_phrase, p_bbox = detect_scale_phrase(words)
        p_factor = scale_factor(p_name)

        for w in words:
            tok = str(w.get("text", ""))
            raw_val = parse_number(tok)
            if raw_val is None:
                continue
            bbox = (
                float(w.get("x0", 0)), float(w.get("top", 0)),
                float(w.get("x1", 0)), float(w.get("bottom", 0))
            )
            y_key = round(float(w.get("top", 0.0)) / 10.0) * 10
            line_words = line_map.get(y_key, [w])

            # Determine region scale
            scale_name = p_name
            scale_phrase = p_phrase
            scale_bbox = p_bbox
            factor = p_factor
            for tb, tn, tp, tbbox in table_scales:
                if inside(bbox, tb):
                    scale_name, scale_phrase, scale_bbox = tn, tp, tbbox
                    factor = scale_factor(tn)
                    break

            # Row/label-based unit classification
            units = classify_units_for_word(w, line_words)

            # Apply scaling only to 'money' or 'unknown' that are not obvious headcount
            apply_factor = factor if units != "people" else 1.0

            hits.append({
                "page_num": page_num,
                "raw": tok,
                "raw_value": raw_val,
                "scaled_value": raw_val * apply_factor,
                "bbox": bbox,
                "units": units,
                "scale_name": scale_name,
                "scale_phrase": scale_phrase,
                "scale_bbox": scale_bbox,
                "table_bbox": next((tb for tb, *_ in table_scales if inside(bbox, tb)), None),
                "page_bbox": page.bbox,
            })

# Apply threshold filters
before = len(hits)

def within_threshold(h):
    if MAX_SCALED is not None and h["scaled_value"] > MAX_SCALED:
        return False
    if MIN_SCALED is not None and h["scaled_value"] < MIN_SCALED:
        return False
    if MAX_RAW is not None and h["raw_value"] > MAX_RAW:
        return False
    if MIN_RAW is not None and h["raw_value"] < MIN_RAW:
        return False
    return True

hits = [h for h in hits if within_threshold(h)]
after = len(hits)
print(f"Thresholds applied — kept {after}/{before} hits "
      f"(MAX_SCALED={MAX_SCALED}, MIN_SCALED={MIN_SCALED}, "
      f"MAX_RAW={MAX_RAW}, MIN_RAW={MIN_RAW})")


# -------- Sort and visualize --------
hits.sort(key=lambda h: h["scaled_value"], reverse=True)
top_hits = hits[:TOP_N]

print(f"Scanned pages {START_PAGE}..{END_PAGE or total_pages}. "
      f"Found {len(hits)} numeric tokens. Showing top {len(top_hits)} by scaled value "
      f"(row-level headcount stays unscaled).")

with pdfplumber.open(pdf_path) as pdf:
    for rank, h in enumerate(top_hits, 1):
        page = pdf.pages[h["page_num"] - 1]
        words = page.extract_words(x_tolerance=X_TOL, y_tolerance=Y_TOL) or []

        print(
            f"#{rank} scaled={h['scaled_value']:,} raw={h['raw_value']} ('{h['raw']}') "
            f"units={h['units']} page={h['page_num']} "
            f"scale={h['scale_name'] or 'none'} phrase={h['scale_phrase']!r} "
            f"bbox={tuple(round(x,1) for x in h['bbox'])}"
        )

        im = page.to_image(resolution=RESOLUTION)

        # Draw tables in green
        try:
            tables = page.find_tables() or []
        except Exception:
            tables = []
        for t in tables:
            if getattr(t, "bbox", None):
                x0, top, x1, bottom = t.bbox
                im.draw_rects([{"x0": x0, "top": top, "x1": x1, "bottom": bottom}],
                              stroke="green", stroke_width=2)

        # Draw page/table scale phrase in blue
        if h["scale_bbox"]:
            x0, top, x1, bottom = h["scale_bbox"]
            im.draw_rects([{"x0": x0, "top": top, "x1": x1, "bottom": bottom}],
                          stroke="blue", stroke_width=2)

        # Draw the number in red
        x0, top, x1, bottom = h["bbox"]
        im.draw_rects([{"x0": x0, "top": top, "x1": x1, "bottom": bottom}],
                      stroke="red", stroke_width=2)

        display(im)
```
